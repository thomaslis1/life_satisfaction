```python
#import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#download and prepare the data

data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
x = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
#visualise the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
#select a linear model
model = LinearRegression()
```

```python
#train the model
model.fit(x, y)
```

```python
# Generate a range of GDP values (x-axis) for the prediction line
x_fit = np.linspace(23_500, 62_500, 100).reshape(-1, 1)

# Predict life satisfaction for those GDP values
y_pred = model.predict(x_fit)

# Plot the original data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")

# Plot the regression line
plt.plot(x_fit, y_pred, color='red', linewidth=2, label='Best fit line')
plt.axis([23_500, 62_500, 4, 9])
plt.legend()
plt.show()
```
